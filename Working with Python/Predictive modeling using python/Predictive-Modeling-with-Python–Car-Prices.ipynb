## Simple Linear Regression

**What it does:** Predicts one thing based on another thing using a straight line.

**Example:** Predict car price based on car age.
- Find the pattern: "Cars lose $2,000 value per year"
- Use the pattern: 5-year-old car = Original price - ($2,000 × 5)

**The process:**
1. Plot data points (age vs. price)
2. Draw the best straight line through the dots
3. Use that line to predict new values

**Bottom line:** Turn data patterns into predictions.

$$
 Y: Response \ Variable\\\\\\
 X: Predictor \ Variables
$$

**The Formula:** `Y = Response Variable, X = Predictor Variable`
- **Y** = What you want to predict (like car price)
- **X** = What you use to predict it (like car mileage)



## Linear Function
$$
Yhat = a + b  X
$$

**The Formula:** `Y = a + bX`

**What each part means:**
- **a** = Starting point (where the line crosses the Y-axis when X = 0)
- **b** = How steep the line is (how much Y goes up for each step of X)
- **X** = What you know (input)
- **Y** = What you're predicting (output)

**Real Example:**
Pizza delivery cost: `Cost = $5 + $2 × Miles`
- **a = $5** → Base delivery fee (even if you live 0 miles away)
- **b = $2** → Extra cost per mile
- **X = Miles** → Distance to your house
- **Y = Cost** → Total delivery cost

So if you live 3 miles away: Cost = $5 + ($2 × 3) = $11

**Bottom Line:** It's just a straight line equation that helps you predict one thing based on another.

## Slope and Intercept  

The **slope** shows how much the output (Y) changes when the input (X) increases by 1 unit.  
- Example: If slope = -150, then for every +1 mpg, the car price goes down by 150.  

The **intercept** is the starting value of Y when X = 0 (where the line crosses the Y-axis).  
- Example: If intercept = 40,000, then at 0 mpg, the predicted car price is 40,000.  

**Formula:**  
$$
Y = (slope × X) + intercept
$$

## **Residual Plot**

A residual plot helps you check if your linear model is working well.

#### What is a residual?
The **residual** = actual price - predicted price

If your model predicts $20,000 but the actual price is $22,000, the residual is +$2,000.

#### What is a residual plot?
A chart that shows:
- **X-axis**: The feature you're using (like highway-mpg)
- **Y-axis**: The residuals (prediction errors)

#### How to read it:
**Good model**: Points are randomly scattered around zero
- No clear pattern
- Points spread evenly above and below the center line

**Bad model**: Points form a pattern (curve, funnel shape, etc.)
- This means a linear model isn't the right fit for your data

#### Bottom line:
Random scatter = linear model works well
Clear pattern = need a different model

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['highway-mpg'], y=df['price'])
plt.show()
```

<i>What is this plot telling us?</i>

<p>We can see from this residual plot that the residuals are not randomly spread around the x-axis, leading us to believe that maybe a non-linear model is more appropriate for this data.</p>

## **Advanced Modeling: Polynomial Regression and Pipelines**

#### Overview
After discovering that linear models had limitations with the highway-mpg data, I explored polynomial regression to capture more complex, non-linear relationships in car pricing.

#### What I Implemented
I moved beyond simple straight-line relationships to curved models that can better represent real-world pricing patterns.

#### Model Types I Used:

**Quadratic (2nd Order):**

$$
Price = a + b₁×mpg + b₂×mpg²
$$


**Cubic (3rd Order):**

$$
Price = a + b₁×mpg + b₂×mpg² + b₃×mpg³
$$


**Higher-Order Models:**

$$
Price = a + b₁×mpg + b₂×mpg² + b₃×mpg³ + ...
$$

#### Why This Matters
Linear models assume the relationship between features and price is constant. Polynomial regression allows me to capture scenarios where:
- The first 10 mpg improvement might save $1000/mpg
- But mpg improvements from 30-40 might only save $200/mpg

This creates more realistic, curved relationships that better match real-world car pricing dynamics.

#### Next Steps
I'll compare polynomial models against linear models to determine which approach provides better predictions for this dataset.

<p>i will use the following function to plot the data:</p>

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

Let's get the variables:


```python
x = df['highway-mpg']
y = df['price']
```

Let's fit the polynomial using the function <b>polyfit</b>, then use the function <b>poly1d</b> to display the polynomial function.


```python
# Here we use a polynomial of the 3rd order (cubic) 
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

Let's plot the function:

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
np.polyfit(x, y, 3)
```

**Polynomial Model Results:**

**3rd order polynomial:** Price = -1.557×mpg³ + 204.8×mpg² - 8965×mpg + 137,900

The curved orange line fits the data much better than a straight line would, capturing the steep price drop at low mpg and leveling off at high mpg.

# Model Evaluation Metrics

## R-squared (R²)
**What it measures:** How well my model explains the data (0 to 1 scale)

- **1.0** = Perfect fit
- **0.5** = Decent fit  
- **0.0** = Terrible fit

**Higher is better**

## Mean Squared Error (MSE)
**What it measures:** Average error in my predictions

- **Lower MSE** = Better predictions
- **Higher MSE** = Worse predictions

**Lower is better**

## Quick Summary
- **R²:** How good is my model? (want high)
- **MSE:** How wrong are my predictions? (want low)
