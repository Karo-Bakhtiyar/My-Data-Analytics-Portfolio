## Individual Data Patterns
### Continuous Variables
Create scatter plots with trend lines for CPU frequency, screen size, and weight versus price. Calculate the correlation between each feature and price.

```python
# Create regression plots and calculate correlations for continuous variables

# CPU Frequency vs Price
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.regplot(x="CPU_frequency", y="Price", data=df)
plt.title("CPU Frequency vs Price")
cpu_corr = df["CPU_frequency"].corr(df["Price"])
print(f"CPU Frequency correlation with Price: {cpu_corr:.4f}")

# Screen Size vs Price
plt.subplot(1, 3, 2)
sns.regplot(x="Screen_Size_inch", y="Price", data=df)
plt.title("Screen Size vs Price")
screen_corr = df["Screen_Size_inch"].corr(df["Price"])
print(f"Screen Size correlation with Price: {screen_corr:.4f}")

# Weight vs Price
plt.subplot(1, 3, 3)
sns.regplot(x="Weight_pounds", y="Price", data=df)
plt.title("Weight vs Price")
weight_corr = df["Weight_pounds"].corr(df["Price"])
print(f"Weight correlation with Price: {weight_corr:.4f}")

plt.tight_layout()
plt.show()
```

Interpretation: "CPU_frequency" has a % positive correlation with the price of the laptops. The other two parameters have weak correlation with price.

## GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


```python
# Group by GPU and CPU_core, calculate mean price
grouped = df.groupby(['GPU', 'CPU_core'])['Price'].agg(['mean', 'count', 'std']).round(2)
print("Grouped by GPU and CPU_core:")
print(grouped)

# Alternative grouping for better understanding
print("\nAverage Price by GPU:")
gpu_prices = df.groupby('GPU')['Price'].mean().round(2).sort_values(ascending=False)
print(gpu_prices)

print("\nAverage Price by CPU_core:")
cpu_prices = df.groupby('CPU_core')['Price'].mean().round(2).sort_values(ascending=False)
print(cpu_prices)
```

```python
# Create pivot table with GPU as rows, CPU_core as columns, Price as values
pivot_table = pd.pivot_table(
    df, 
    values='Price', 
    index='GPU', 
    columns='CPU_core', 
    aggfunc='mean'  # Calculate mean price
)

print("Pivot Table (GPU vs CPU_core vs Price):")
print(pivot_table.round(2))

# Fill NaN values with 0 for visualization
pivot_filled = pivot_table.fillna(0)
print("\nPivot Table (NaN filled with 0):")
print(pivot_filled.round(2))
```

```python
# Create pcolor plot
plt.figure(figsize=(10, 6))
plt.pcolor(pivot_filled.values, cmap='viridis')

# Add colorbar
plt.colorbar(label='Average Price ($)')

# Set labels and title
plt.title('Price by GPU and CPU Cores (Pcolor Plot)', fontsize=14, fontweight='bold')
plt.xlabel('CPU Cores', fontsize=12)
plt.ylabel('GPU Type', fontsize=12)

# Set tick labels
plt.xticks(np.arange(len(pivot_filled.columns)) + 0.5, pivot_filled.columns)
plt.yticks(np.arange(len(pivot_filled.index)) + 0.5, pivot_filled.index)

# Rotate y-axis labels for better readability
plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

# Pearson Correlation Analysis for Laptop Pricing

Used `scipy.stats.pearsonr()` to identify which laptop specifications (CPU, RAM, storage, screen size) most strongly correlate with price. Statistical analysis revealed key pricing drivers for business strategy.

**Skills:** Correlation analysis, statistical significance testing, business insights.

```python
from scipy.stats import pearsonr

print("PEARSON CORRELATION WITH PRICE")
print("=" * 35)

# Analyze key features
features = ['CPU_core', 'RAM_GB', 'Storage_GB_SSD', 'Screen_Size_inch']

for feature in features:
    corr, p_val = pearsonr(df[feature], df['Price'])
    significance = "***" if p_val < 0.001 else "**" if p_val < 0.01 else "*" if p_val < 0.05 else ""
    print(f"{feature:18}: r={corr:6.3f}, p={p_val:.3f} {significance}")

print("\n* p<0.05, ** p<0.01, *** p<0.001")
```

Strong correlations (all statistically significant):

RAM (r=0.549): Strongest price driver
CPU cores (r=0.459): Second strongest
Storage (r=0.243): Moderate impact

Screen size (r=-0.111): Very weak negative correlation, NOT statistically significant (no stars)
Bottom line: RAM and CPU cores strongly drive price up. Storage has some impact. Screen size has virtually no reliable relationship with price.
